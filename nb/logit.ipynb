```python
import numpy as np
import matplotlib.pyplot as plt

def plot_logistic_transformation():
    # Generate probabilities from 0.001 to 0.999
    p = np.linspace(0.001, 0.999, 1000)
    
    # Calculate odds and log-odds
    odds = p / (1 - p)
    log_odds = np.log(odds)
    
    # Create the visualization
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))
    
    # Plot probability
    ax1.plot(p, p)
    ax1.set_title('Probability')
    ax1.set_ylim(-0.1, 1.1)
    
    # Plot odds
    ax2.plot(p, odds)
    ax2.set_title('Odds')
    ax2.set_ylim(-1, 10)
    
    # Plot log-odds
    ax3.plot(p, log_odds)
    ax3.set_title('Log-odds')
    ax3.axhline(y=0, color='r', linestyle='--')
    
    plt.tight_layout()
    return fig

plot_logistic_transformation()
```

```python
def explain_logistic_function():
    # Start with log-odds = x
    # log(p/(1-p)) = x
    # p/(1-p) = exp(x)
    # p = (1-p)exp(x)
    # p = exp(x) - p*exp(x)
    # p + p*exp(x) = exp(x)
    # p(1 + exp(x)) = exp(x)
    # p = exp(x)/(1 + exp(x))
    # This is equivalent to:
    # p = 1/(1 + exp(-x))
    
    # Let's visualize this
    x = np.linspace(-6, 6, 100)
    y = 1 / (1 + np.exp(-x))
    
    plt.figure(figsize=(10, 5))
    plt.plot(x, y)
    plt.title('Sigmoid Function')
    plt.grid(True)
    plt.axhline(y=0.5, color='r', linestyle='--')
    plt.axvline(x=0, color='r', linestyle='--')
    return plt.gcf()

explain_logistic_function()
```
